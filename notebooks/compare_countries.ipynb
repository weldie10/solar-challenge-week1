## 2. Metric Comparison

### 2.1 Boxplots: GHI, DNI, DHI Side-by-Side


```python
def create_comparison_boxplots(df_combined, metrics=['GHI', 'DNI', 'DHI']):
    """Create side-by-side boxplots for solar metrics across countries"""
    
    # Filter available metrics
    available_metrics = [m for m in metrics if m in df_combined.columns]
    
    if not available_metrics:
        print("‚ö†Ô∏è No metrics available for comparison")
        return
    
    # Create subplots
    n_metrics = len(available_metrics)
    fig, axes = plt.subplots(1, n_metrics, figsize=(6*n_metrics, 6))
    
    if n_metrics == 1:
        axes = [axes]
    
    # Color palette for countries
    colors = {'Benin': '#2E86AB', 'Sierra Leone': '#A23B72', 'Togo': '#F18F01'}
    
    for idx, metric in enumerate(available_metrics):
        ax = axes[idx]
        
        # Prepare data for boxplot
        data_for_plot = []
        labels = []
        plot_colors = []
        
        for country in df_combined['Country'].unique():
            country_data = df_combined[df_combined['Country'] == country][metric].dropna()
            if len(country_data) > 0:
                data_for_plot.append(country_data)
                labels.append(country)
                plot_colors.append(colors.get(country, '#808080'))
        
        # Create boxplot
        bp = ax.boxplot(data_for_plot, labels=labels, patch_artist=True, 
                        showmeans=True, meanline=False)
        
        # Color the boxes
        for patch, color in zip(bp['boxes'], plot_colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)
        
        # Customize plot
        ax.set_title(f'{metric} Comparison Across Countries', fontsize=12, fontweight='bold')
        ax.set_ylabel(f'{metric} (W/m¬≤)', fontsize=10)
        ax.set_xlabel('Country', fontsize=10)
        ax.grid(True, alpha=0.3, axis='y')
        
    plt.tight_layout()
    plt.show()
    
    print(f"\n‚úÖ Boxplots created for: {', '.join(available_metrics)}")

# Create boxplots
if 'df_combined' in locals():
    create_comparison_boxplots(df_combined)
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```

### 2.2 Summary Table: Mean, Median, and Standard Deviation


```python
def create_summary_table(df_combined, metrics=['GHI', 'DNI', 'DHI']):
    """Create comprehensive summary table comparing countries"""
    
    available_metrics = [m for m in metrics if m in df_combined.columns]
    
    if not available_metrics:
        print("‚ö†Ô∏è No metrics available for summary")
        return None
    
    # Initialize summary dictionary
    summary_data = []
    
    for country in sorted(df_combined['Country'].unique()):
        country_df = df_combined[df_combined['Country'] == country]
        
        for metric in available_metrics:
            metric_data = country_df[metric].dropna()
            
            if len(metric_data) > 0:
                summary_data.append({
                    'Country': country,
                    'Metric': metric,
                    'Mean': metric_data.mean(),
                    'Median': metric_data.median(),
                    'Std Dev': metric_data.std(),
                    'Min': metric_data.min(),
                    'Max': metric_data.max(),
                    'Count': len(metric_data)
                })
    
    # Create DataFrame
    summary_df = pd.DataFrame(summary_data)
    
    # Format for display
    print("\n" + "="*80)
    print("SUMMARY STATISTICS: Solar Metrics Comparison Across Countries")
    print("="*80)
    
    # Display by metric
    for metric in available_metrics:
        metric_df = summary_df[summary_df['Metric'] == metric].copy()
        if len(metric_df) > 0:
            print(f"\nüìä {metric} Statistics:")
            print("-"*80)
            display_df = metric_df[['Country', 'Mean', 'Median', 'Std Dev', 'Min', 'Max', 'Count']].copy()
            display_df['Mean'] = display_df['Mean'].round(2)
            display_df['Median'] = display_df['Median'].round(2)
            display_df['Std Dev'] = display_df['Std Dev'].round(2)
            display_df['Min'] = display_df['Min'].round(2)
            display_df['Max'] = display_df['Max'].round(2)
            print(display_df.to_string(index=False))
    
    return summary_df

# Create summary table
if 'df_combined' in locals():
    summary_table = create_summary_table(df_combined)
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```

## 3. Statistical Testing

### 3.1 One-Way ANOVA Test for GHI


```python
def perform_anova_test(df_combined, metric='GHI'):
    """Perform one-way ANOVA test to assess differences between countries"""
    
    if metric not in df_combined.columns:
        print(f"‚ö†Ô∏è Metric '{metric}' not found in dataset")
        return None
    
    # Prepare data for ANOVA
    groups = []
    group_names = []
    
    for country in sorted(df_combined['Country'].unique()):
        country_data = df_combined[df_combined['Country'] == country][metric].dropna()
        if len(country_data) > 0:
            groups.append(country_data.values)
            group_names.append(country)
    
    if len(groups) < 2:
        print(f"‚ö†Ô∏è Need at least 2 countries for ANOVA test")
        return None
    
    print(f"\n" + "="*80)
    print(f"STATISTICAL TESTING: {metric} Comparison Across Countries")
    print("="*80)
    
    # Perform ANOVA
    print("\nüìä One-Way ANOVA Test:")
    print("-"*80)
    
    try:
        f_statistic, p_value_anova = f_oneway(*groups)
        
        print(f"F-statistic: {f_statistic:.4f}")
        print(f"p-value: {p_value_anova:.6f}")
        
        alpha = 0.05
        if p_value_anova < alpha:
            print(f"\n‚úÖ Result: Significant difference detected (p < {alpha})")
            print(f"   ‚Üí There is statistically significant difference in {metric} between countries")
        else:
            print(f"\n‚ùå Result: No significant difference (p ‚â• {alpha})")
            print(f"   ‚Üí No statistically significant difference in {metric} between countries")
        
        anova_result = {
            'test': 'One-Way ANOVA',
            'metric': metric,
            'f_statistic': f_statistic,
            'p_value': p_value_anova,
            'significant': p_value_anova < alpha
        }
        
    except Exception as e:
        print(f"‚ö†Ô∏è ANOVA test failed: {e}")
        anova_result = None
    
    # Perform Kruskal-Wallis test (non-parametric alternative)
    print("\nüìä Kruskal-Wallis Test (Non-parametric Alternative):")
    print("-"*80)
    
    try:
        h_statistic, p_value_kw = kruskal(*groups)
        
        print(f"H-statistic: {h_statistic:.4f}")
        print(f"p-value: {p_value_kw:.6f}")
        
        alpha = 0.05
        if p_value_kw < alpha:
            print(f"\n‚úÖ Result: Significant difference detected (p < {alpha})")
            print(f"   ‚Üí There is statistically significant difference in {metric} between countries")
        else:
            print(f"\n‚ùå Result: No significant difference (p ‚â• {alpha})")
            print(f"   ‚Üí No statistically significant difference in {metric} between countries")
        
        kw_result = {
            'test': 'Kruskal-Wallis',
            'metric': metric,
            'h_statistic': h_statistic,
            'p_value': p_value_kw,
            'significant': p_value_kw < alpha
        }
        
        return {'anova': anova_result, 'kruskal_wallis': kw_result}
        
    except Exception as e:
        print(f"‚ö†Ô∏è Kruskal-Wallis test failed: {e}")
        return {'anova': anova_result, 'kruskal_wallis': None}

# Perform statistical tests for GHI
if 'df_combined' in locals():
    test_results = perform_anova_test(df_combined, metric='GHI')
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```

### 3.2 Post-Hoc Analysis: Pairwise Comparisons

To identify which specific countries differ significantly, we perform post-hoc pairwise comparisons using Tukey's HSD test.


```python
from scipy.stats import tukey_hsd
from itertools import combinations

def perform_pairwise_comparisons(df_combined, metric='GHI'):
    """Perform pairwise comparisons between countries using statistical tests"""
    
    if metric not in df_combined.columns:
        print(f"‚ö†Ô∏è Metric '{metric}' not found in dataset")
        return None
    
    # Prepare data for pairwise comparison
    groups_dict = {}
    for country in sorted(df_combined['Country'].unique()):
        country_data = df_combined[df_combined['Country'] == country][metric].dropna()
        if len(country_data) > 0:
            groups_dict[country] = country_data.values
    
    if len(groups_dict) < 2:
        print(f"‚ö†Ô∏è Need at least 2 countries for pairwise comparison")
        return None
    
    print(f"\n" + "="*80)
    print(f"PAIRWISE COMPARISONS: {metric} Between Countries")
    print("="*80)
    
    # Get all country pairs
    countries = sorted(groups_dict.keys())
    pairs = list(combinations(countries, 2))
    
    results = []
    
    # Use Tukey HSD with all groups at once for proper multiple comparisons correction
    try:
        all_groups = [groups_dict[country] for country in countries]
        tukey_result = tukey_hsd(*all_groups)
        
        # Extract p-value matrix (symmetric matrix)
        pvalue_matrix = tukey_result.pvalue
        
        for idx, (country1, country2) in enumerate(pairs):
            # Get indices in the countries list
            idx1 = countries.index(country1)
            idx2 = countries.index(country2)
            
            # Extract p-value from matrix (matrix is symmetric)
            p_value = pvalue_matrix[idx1, idx2]
            
            # Calculate mean difference
            mean_diff = groups_dict[country1].mean() - groups_dict[country2].mean()
            
            # Determine significance
            is_significant = p_value < 0.05
            
            results.append({
                'Country_1': country1,
                'Country_2': country2,
                'Mean_Diff': mean_diff,
                'P_Value': p_value,
                'Significant': is_significant
            })
            
            significance = "‚úÖ Significant" if is_significant else "‚ùå Not Significant"
            print(f"\n{country1} vs {country2}:")
            print(f"  Mean Difference: {mean_diff:.2f} W/m¬≤")
            print(f"  p-value: {p_value:.6f}")
            print(f"  Result: {significance}")
            
    except Exception as e:
        print(f"‚ö†Ô∏è Error in Tukey HSD test: {e}")
        print("Falling back to individual t-tests...")
        
        # Fallback to t-tests for pairwise comparisons
        from scipy.stats import ttest_ind
        
        for country1, country2 in pairs:
            group1 = groups_dict[country1]
            group2 = groups_dict[country2]
            
            try:
                # Perform independent t-test
                t_stat, p_value = ttest_ind(group1, group2)
                
                # Calculate mean difference
                mean_diff = group1.mean() - group2.mean()
                
                # Determine significance
                is_significant = p_value < 0.05
                
                results.append({
                    'Country_1': country1,
                    'Country_2': country2,
                    'Mean_Diff': mean_diff,
                    'P_Value': p_value,
                    'Significant': is_significant
                })
                
                significance = "‚úÖ Significant" if is_significant else "‚ùå Not Significant"
                print(f"\n{country1} vs {country2} (t-test):")
                print(f"  Mean Difference: {mean_diff:.2f} W/m¬≤")
                print(f"  p-value: {p_value:.6f}")
                print(f"  Result: {significance}")
                
            except Exception as e2:
                print(f"‚ö†Ô∏è Error comparing {country1} vs {country2}: {e2}")
    
    if results:
        results_df = pd.DataFrame(results)
        
        print("\n" + "-"*80)
        print("Summary of Pairwise Comparisons:")
        print("-"*80)
        display_df = results_df.copy()
        display_df['P_Value'] = display_df['P_Value'].round(6)
        display_df['Mean_Diff'] = display_df['Mean_Diff'].round(2)
        print(display_df.to_string(index=False))
        
        return results_df
    else:
        print("\n‚ö†Ô∏è No pairwise comparisons completed")
        return None

# Perform pairwise comparisons for GHI
if 'df_combined' in locals():
    pairwise_results = perform_pairwise_comparisons(df_combined, metric='GHI')
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```

### 3.3 Additional Visualizations: Violin Plots

Violin plots provide a more detailed view of the distribution shape compared to boxplots, showing the probability density of the data.


```python
def create_violin_plots(df_combined, metrics=['GHI', 'DNI', 'DHI']):
    """Create violin plots for better distribution visualization"""
    
    available_metrics = [m for m in metrics if m in df_combined.columns]
    
    if not available_metrics:
        print("‚ö†Ô∏è No metrics available for comparison")
        return
    
    # Create subplots
    n_metrics = len(available_metrics)
    fig, axes = plt.subplots(1, n_metrics, figsize=(6*n_metrics, 6))
    
    if n_metrics == 1:
        axes = [axes]
    
    # Color palette for countries
    colors = {'Benin': '#2E86AB', 'Sierra Leone': '#A23B72', 'Togo': '#F18F01'}
    
    for idx, metric in enumerate(available_metrics):
        ax = axes[idx]
        
        # Create violin plot
        data_list = []
        labels_list = []
        
        for country in sorted(df_combined['Country'].unique()):
            country_data = df_combined[df_combined['Country'] == country][metric].dropna()
            if len(country_data) > 0:
                data_list.append(country_data.values)
                labels_list.append(country)
        
        # Create violin plot
        parts = ax.violinplot(data_list, positions=range(len(labels_list)), 
                              showmeans=True, showmedians=True)
        
        # Color the violins
        for pc, country in zip(parts['bodies'], labels_list):
            pc.set_facecolor(colors.get(country, '#808080'))
            pc.set_alpha(0.7)
        
        # Customize plot
        ax.set_xticks(range(len(labels_list)))
        ax.set_xticklabels(labels_list)
        ax.set_title(f'{metric} Distribution Comparison', fontsize=12, fontweight='bold')
        ax.set_ylabel(f'{metric} (W/m¬≤)', fontsize=10)
        ax.set_xlabel('Country', fontsize=10)
        ax.grid(True, alpha=0.3, axis='y')
        
    plt.tight_layout()
    plt.show()
    
    print(f"\n‚úÖ Violin plots created for: {', '.join(available_metrics)}")

# Create violin plots
if 'df_combined' in locals():
    create_violin_plots(df_combined)
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```

### 3.4 Data Quality and Completeness Check

Before drawing conclusions, we verify data quality and completeness across all countries.


```python
def check_data_quality(df_combined, metrics=['GHI', 'DNI', 'DHI']):
    """Comprehensive data quality check across countries"""
    
    print("\n" + "="*80)
    print("DATA QUALITY AND COMPLETENESS REPORT")
    print("="*80)
    
    quality_report = []
    
    for country in sorted(df_combined['Country'].unique()):
        country_df = df_combined[df_combined['Country'] == country]
        
        for metric in metrics:
            if metric in country_df.columns:
                metric_data = country_df[metric]
                
                quality_report.append({
                    'Country': country,
                    'Metric': metric,
                    'Total_Records': len(metric_data),
                    'Non_Null': metric_data.notna().sum(),
                    'Null_Count': metric_data.isna().sum(),
                    'Null_Percentage': (metric_data.isna().sum() / len(metric_data) * 100),
                    'Negative_Values': (metric_data < 0).sum() if metric_data.notna().any() else 0,
                    'Zero_Values': (metric_data == 0).sum() if metric_data.notna().any() else 0
                })
    
    quality_df = pd.DataFrame(quality_report)
    
    print("\nüìä Data Completeness by Country and Metric:")
    print("-"*80)
    display_cols = ['Country', 'Metric', 'Total_Records', 'Non_Null', 'Null_Percentage']
    print(quality_df[display_cols].to_string(index=False))
    
    print("\nüìä Data Quality Issues:")
    print("-"*80)
    issues = quality_df[(quality_df['Null_Percentage'] > 0) | 
                        (quality_df['Negative_Values'] > 0) |
                        (quality_df['Zero_Values'] > 1000)]
    
    if len(issues) > 0:
        print(issues[['Country', 'Metric', 'Null_Percentage', 'Negative_Values', 'Zero_Values']].to_string(index=False))
    else:
        print("‚úÖ No significant data quality issues detected")
    
    return quality_df

# Perform data quality check
if 'df_combined' in locals():
    quality_report = check_data_quality(df_combined)
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```

### 5.2 Export Results Summary

Export key results to CSV for further analysis or reporting.


```python
def export_results_summary(summary_table, test_results, output_dir='../results'):
    """Export analysis results to CSV files"""
    
    from pathlib import Path
    import os
    
    output_path = Path(output_dir)
    output_path.mkdir(exist_ok=True)
    
    # Export summary statistics
    if summary_table is not None:
        summary_file = output_path / 'cross_country_summary_statistics.csv'
        summary_table.to_csv(summary_file, index=False)
        print(f"‚úÖ Summary statistics exported to: {summary_file}")
    
    # Export statistical test results
    if test_results:
        test_results_list = []
        for metric, results in test_results.items():
            if results and 'anova' in results and results['anova']:
                anova = results['anova']
                test_results_list.append({
                    'Metric': metric,
                    'Test': 'ANOVA',
                    'Statistic': anova.get('f_statistic', None),
                    'P_Value': anova.get('p_value', None),
                    'Significant': anova.get('significant', False)
                })
            if results and 'kruskal_wallis' in results and results['kruskal_wallis']:
                kw = results['kruskal_wallis']
                test_results_list.append({
                    'Metric': metric,
                    'Test': 'Kruskal-Wallis',
                    'Statistic': kw.get('h_statistic', None),
                    'P_Value': kw.get('p_value', None),
                    'Significant': kw.get('significant', False)
                })
        
        if test_results_list:
            test_results_df = pd.DataFrame(test_results_list)
            test_file = output_path / 'cross_country_statistical_tests.csv'
            test_results_df.to_csv(test_file, index=False)
            print(f"‚úÖ Statistical test results exported to: {test_file}")
    
    print(f"\nüìÅ All results exported to: {output_path.absolute()}")

# Export results
if 'summary_table' in locals() and 'test_results' in locals():
    export_results_summary(summary_table, test_results)
else:
    print("‚ö†Ô∏è Results not available for export. Please run analysis first.")
```

```python
def create_ghi_ranking_chart(df_combined):
    """Create bar chart ranking countries by average GHI"""
    
    if 'GHI' not in df_combined.columns:
        print("‚ö†Ô∏è GHI metric not available")
        return
    
    # Calculate average GHI by country
    ghi_by_country = df_combined.groupby('Country')['GHI'].agg(['mean', 'median', 'std']).reset_index()
    ghi_by_country = ghi_by_country.sort_values('mean', ascending=False)
    
    # Create figure with subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # Color palette
    colors = {'Benin': '#2E86AB', 'Sierra Leone': '#A23B72', 'Togo': '#F18F01'}
    bar_colors = [colors.get(country, '#808080') for country in ghi_by_country['Country']]
    
    # Plot 1: Mean GHI ranking
    bars1 = ax1.bar(ghi_by_country['Country'], ghi_by_country['mean'], 
                    color=bar_colors, alpha=0.8, edgecolor='black', linewidth=1.5)
    
    # Add value labels on bars
    for i, (country, mean_val) in enumerate(zip(ghi_by_country['Country'], ghi_by_country['mean'])):
        ax1.text(i, mean_val + max(ghi_by_country['mean']) * 0.01, 
                f'{mean_val:.1f}', ha='center', va='bottom', fontweight='bold', fontsize=10)
    
    ax1.set_title('Countries Ranked by Average GHI', fontsize=14, fontweight='bold')
    ax1.set_ylabel('Average GHI (W/m¬≤)', fontsize=11)
    ax1.set_xlabel('Country', fontsize=11)
    ax1.grid(True, alpha=0.3, axis='y')
    ax1.set_ylim(0, max(ghi_by_country['mean']) * 1.15)
    
    # Add ranking numbers
    for i, country in enumerate(ghi_by_country['Country']):
        ax1.text(i, -max(ghi_by_country['mean']) * 0.08, f'#{i+1}', 
                ha='center', va='top', fontsize=12, fontweight='bold', color='gray')
    
    # Plot 2: Mean vs Median comparison
    x = np.arange(len(ghi_by_country))
    width = 0.35
    
    bars2 = ax2.bar(x - width/2, ghi_by_country['mean'], width, 
                    label='Mean', color=bar_colors, alpha=0.8, edgecolor='black')
    bars3 = ax2.bar(x + width/2, ghi_by_country['median'], width, 
                    label='Median', color=bar_colors, alpha=0.5, edgecolor='black')
    
    ax2.set_title('Mean vs Median GHI Comparison', fontsize=14, fontweight='bold')
    ax2.set_ylabel('GHI (W/m¬≤)', fontsize=11)
    ax2.set_xlabel('Country', fontsize=11)
    ax2.set_xticks(x)
    ax2.set_xticklabels(ghi_by_country['Country'])
    ax2.legend()
    ax2.grid(True, alpha=0.3, axis='y')
    
    plt.tight_layout()
    plt.show()
    
    # Print ranking table
    print("\n" + "="*80)
    print("COUNTRY RANKING BY AVERAGE GHI")
    print("="*80)
    print("\nRank | Country        | Mean GHI (W/m¬≤) | Median GHI (W/m¬≤) | Std Dev")
    print("-"*80)
    
    for rank, (idx, row) in enumerate(ghi_by_country.iterrows(), 1):
        print(f"{rank:4d} | {row['Country']:14s} | {row['mean']:15.2f} | {row['median']:17.2f} | {row['std']:7.2f}")
    
    print("\n‚úÖ Ranking chart created successfully!")

# Create ranking chart
if 'df_combined' in locals():
    create_ghi_ranking_chart(df_combined)
else:
    print("‚ö†Ô∏è Combined dataset not available. Please load data first.")
```
